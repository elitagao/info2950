```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import re 
from scipy import stats

import seaborn as sns
```

```python
combined_ulta_df= pd.read_csv('combined_ulta.csv')
combined_ulta_df.head()
```

#### The following code cleans prices...

```python

step=-1
for x in combined_ulta_df['price']:
    step+=1
    if "Price\n$" in x:
        find_word=combined_ulta_df.loc[step,'price'].find('Price\n')
        combined_ulta_df.loc[step,'price']= combined_ulta_df.loc[step,'price'][find_word+7:]

            
combined_ulta_df.head()
```

```python
combined_ulta_df.head(25)
```

```python
step=-1
for x in combined_ulta_df['price']:
    step+=1
    if "Original" in x:
        find_original=combined_ulta_df.loc[step,'price'].find('Original')
        combined_ulta_df.loc[step,'price']= combined_ulta_df.loc[step,'price'][find_original+16:]

            
combined_ulta_df.head()
```

```python
step=-1
for x in combined_ulta_df['price']:
    step+=1
    if "Market" in x:
        find_market=combined_ulta_df.loc[step,'price'].find('Market')
        combined_ulta_df.loc[step,'price']= combined_ulta_df.loc[step,'price'][:find_market-1]

            



```
